```python
#Tính g(f(căn 3)) , vẽ 2 đồ thị trên 2 biểu đồ khác nhau , cho hiển thị điểm đó trên g(X) với fx=3**x+2, gx=căn 3(x)
def f(x):
    return pow(3, x)+2
def g(x):
    return pow(x, 1/3)
m=f(pow(3,1/2))
n=g(m)
print("g(f(pow(3,1/2)))=(",m,", ",n,")")

import numpy as np
import matplotlib.pyplot as a
fig, (ax1, ax2) = a.subplots(2, 1)

x1=np.linspace(1,10,100)
fx1=pow(3, x)+2
ax1.plot(x1,fx1)
ax1.grid(True)

x2=np.linspace(-10,10,100)
fx2=pow(x, 1/3)
ax2.plot(x2, fx2)
ax2.scatter(m, n)
ax2.grid(True)

a.tight_layout()
a.show 
```

```python
# #vẽ 2 đồ thị trên 1 biểu đồ với fx=3**x+2, gx=căn 3(x)
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(1, 10, 100)  # Giới hạn phạm vi của x1 để tránh tràn số
fx1 = 3**x1 + 2

x2 = np.linspace(1, 100, 1000)  # Tăng phạm vi của x2 để hiển thị đúng hàm căn bậc 3
fx2 = np.cbrt(x2)

# Vẽ đồ thị
graph1, = plt.plot(x1, fx1, label=r'$f(x) = 3^x + 2$')
graph2, = plt.plot(x2, fx2, label=r'$g(x) = \sqrt[3]{x}$')

# Hiển thị chú thích
plt.legend(handles=(graph1, graph2))

# Hiển thị đồ thị
plt.grid(True)
plt.show()
```

```python
#Vẽ hàm arctan(x)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = np.arctan(x)
plt.plot(x, y, label=r'$y = \arctan(x)$')

# Hiển thị chú thích
plt.legend()

# Hiển thị đồ thị
plt.grid(True)
plt.show()
```

```python
# Cho 2 ma trận vuông cấp 2 A=[[1  -2][-4  1]] B=[[1  -8][-7  3]] 
# Tính A+B, A-B, (1/2)*A, AB, BA, nếu Định thức |A|#0 thì tìm ma trận nghịch đảo A-1 
import numpy as np
A=np.array([[1, -2], [-4, 1]])
B=np.array([[1, -8], [-7, 3]])
print("A+B=\n", A+B)
print("A-B=\n", A-B)
print("(1/2)*A=\n", (1/2)*A)
print("AB=\n", np.dot(A,B))
print("BA=\n", np.dot(B, A))
if (np.linalg.det(A))!=0:
    print("A-1=\n", np.linalg.inv(A))
```
