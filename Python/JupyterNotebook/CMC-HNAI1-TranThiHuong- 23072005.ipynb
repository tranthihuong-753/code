```python
#bai 1
import numpy as np
import matplotlib.pyplot as a
z=np.linspace(-8, 8, 100)
y=1 / (1 + np.exp(-z))
a.plot(z,y, label=r'$y = \frac{1}{1 + e^x}$')
a.legend()
a.grid(True)
a.show()
```

```python
# bi 2
import sympy as sp 
x=sp.Symbol('x')
y=pow(x, 3)-3*pow(x, 2)+2*x 
print("f'(x)=", sp.diff(y,x) ) 
#V  th 
import numpy as np
import matplotlib.pyplot as a
x=np.linspace(-10, 10, 100)
y=pow(x, 3)-3*pow(x, 2)+2*x 
a.plot(x,y)
a.grid(True)
a.show()
```
