```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
z = np.linspace(-8, 8, 400)
sigma_z = sigmoid(z)
plt.plot(z, sigma_z)
plt.title(r'$\sigma(z) = \frac{1}{1 + e^{-z}}$')
plt.xlabel('z')
plt.ylabel(r'$\sigma(z)$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x**2 + 2*x

def df(x):
    return 3*x**2 - 6*x + 2

x = np.linspace(-10, 10, 400)
y = df(x)
plt.plot(x, y, label=r"$f'(x) = 3x^2 - 6x + 2$")
plt.title(r"Derivative of $f(x) = x^3 - 3x^2 + 2x$")
plt.xlabel('x')
plt.ylabel(r"$f'(x)$")
plt.show()
```
